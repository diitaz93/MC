```python
t=arange(0,5,0.2)
figure()
axis([0,5,0,120])
plot(t,t,'r--')
plot(t,t**2,'bs')
plot(t,t**3,'g^')
show()
```

```python
t=linspace(0,5,1000)
t2=linspace(0,5,50)
figure()
subplot(2,1,1)
axis([0,5,-1,1])
plot(t,exp(-t)*cos(2*pi*t),'k')
plot(t2,exp(-t2)*cos(2*pi*t2),'bo')

subplot(2,1,2)
axis([0,5,-1,1])
plot(t,cos(2*pi*t),'r--')
show()
```
