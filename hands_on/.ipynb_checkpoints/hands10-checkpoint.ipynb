```python
%pylab inline
from scipy.fftpack import ifft, fft, fftfreq
from scipy import signal
```

```python
data=genfromtxt("solar.csv",delimiter=",")
year=[]
month=[]
spot=[]
```

```python
for i in range(len(data)/2,len(data)-1):
    spot.append(data[i][4])
    year.append(data[i][0])
    month.append(data[i][1])
```

```python
mes=month
for i in range(len(month)):
    mes[i]=month[i]/13
```

```python
time=[]
for i in range(len(year)):
    time.append(year[i]+mes[i])
```

```python
plot(time,spot)
```

```python
N=len(spot)
f=12.0
dt=1./f/N
freq=fftfreq(N,dt)
trans=fft(spot)
plot(freq,abs(trans))
trans[abs(freq) > 300] = 0
filt=ifft(trans)
```

```python
figure(figsize=(12,9))
plot(time, real(filt),'r',lw=2)
plot(time,spot,'b',alpha=0.3)
xlim(1900,2000)
ylabel(r"$Manchas solares/mes$",fontsize=20)
xlabel(ur"$Tiempo (años)$",fontsize=20)
```
