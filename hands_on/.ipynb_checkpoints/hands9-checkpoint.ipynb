#Lena

```python
lena=misc.lena()
inv_lena=255-lena
lena_gh=zeros((len(lena),len(lena)))
lena_gv=zeros((len(lena),len(lena)))
for i in range(len(lena)):
    for j in range(len(lena)):
        lena_gh[i][len(lena)-1-j]=lena[i][j]
        lena_gv[len(lena)-1-i][j]=lena[i][j]
figure(figsize=(10,10))
subplot(2,2,1)
imshow(lena)
subplot(2,2,2)
imshow(lena_gh,cmap=cm.gray)
subplot(2,2,3)
imshow(lena_gv,cmap=cm.gray)
subplot(2,2,4)
imshow(inv_lena, cmap=cm.gray)
show()
```
