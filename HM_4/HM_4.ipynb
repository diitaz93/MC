```python
%pylab inline
from scipy import interpolate
from scipy.optimize import fminbound
```

#a)

```python
#Se crea un arreglo bidimensional de dimensiones 1024x1024 inicializado en cero por practicidad
a=[[0 for i in range(1024)] for i in range(1024)]
#Se crean variables para facilitar la creacion de unos en el centro
lado=32 #Lado del cuadrado central de unos
liminf=len(a)/2-lado/2-1 #Límite inferior
limsup=len(a)/2+lado/2-1 #Límite superior
```

```python
#Se reemplazan los valores del cuadrado cental con unos
for i in range(liminf,limsup):
    for j in range(liminf,limsup):
        a[i][j]=1
#Se grafican en escala de grises: Negro es 0 y blanco es 1.
figure(figsize=(5,5))
imshow(a,cmap=cm.gray)
show()
```

#b)

```python
figure(figsize=(10,6))
#Se realiza la transformada de fourier en 2-D para el arreglo crado anteriormente
trans=fft2(a)
subplot(1,2,1)
#Se muestra la transformada directa y...
imshow(abs(trans))
title('Transformada de Fourier')
trans=fftshift(trans)
subplot(1,2,2)
#Se muestra la transformada centrada en cero
title('Transformada centrada en cero')
imshow(abs(trans))
show()
```

```python
#Se definen nuevas variables para el cuadrado a tratar
lado2=200
liminf=len(a)/2-lado2/2-1
limsup=len(a)/2+lado2/2-1
```

```python
#Se crea un nuevo arreglo bidimensional de tipo complejo con las dimensiones del nuevo cuadrado
b=zeros((lado2,lado2),complex)
#Se llena el nuevo arreglo con los valores deseados de la transformada centrada en cero
for i in range(lado2):
    for j in range(lado2):
        b[i][j]=trans[i+liminf][j+liminf]
imshow(abs(b))
```

```python
#Se crea el arreglo del logaritmo del modulo al cuadrado
final=(log((abs(b))**2))
#Y se grafica junto con la apertura correspondiente
figure(figsize=(15,9))
subplot(1,2,1)
imshow(a,cmap=cm.gray)
subplot(1,2,2)
title(u'Patrón de interferencia de una apertura cuadrada \n (log resaltado)')
imshow(final)
```

```python
#Se crea un arreglo bidimensional de dimensiones 1024x1024 inicializado en cero por practicidad
c1=[[0 for i in range(1024)] for i in range(1024)]
#Se crean variables para facilitar la creación de las rendijas
alto=200
ancho=20
ej1=460
ej2=564
liminf=len(c1)/2-alto/2-1
limsup=len(c1)/2+alto/2-1
limi1=ej1-ancho/2-1
limd1=ej1+ancho/2-1
limi2=ej2-ancho/2-1
limd2=ej2+ancho/2-1
#Se crea el arreglo con la doble rendija representada por unos en una matriz de ceros
for i in range(liminf,limsup):
    for j in range(limi1,limd1):
        c1[i][j]=1
    for k in range(limi2,limd2):
        c1[i][k]=1
#Segrafica a blanco y negro
imshow(c1,cmap=cm.gray)
```

```python
#Se hace la transformada de fourier para la doble rendija
trans=fft2(c1)
figure(figsize=(10,6))
subplot(1,2,1)
title('Transformada de Fourier')
imshow(abs(trans))
#Y se grafica junto con la transformada centrada en cero
trans=fftshift(trans)
subplot(1,2,2)
title('Transformada centrada en cero')
imshow(abs(trans))
```

```python
#Se crean las variables necesarias para seleccionar un cuadrado especifico del arreglo
lado=128
c2=zeros((lado,lado),complex)
liminf=len(c1)/2-lado/2-1
limsup=len(c1)/2+lado/2-1
#Se llena el nuevo arreglo con los valores deseados de la transformada centrada en cero
for i in range(lado):
    for j in range(lado):
        c2[i][j]=trans[i+liminf][j+liminf]
#Se grafica la doble rendija junto con la transformada centrada en cero en un cuadrado específico
figure(figsize=(15,9))
subplot(1,2,1)
imshow(c1,cmap=cm.gray)
subplot(1,2,2)
title(u'Patrón de difracción de una doble rendija')
imshow(abs(c2))
show()
```

```python
# Para realizar la gráfica superior (intesidad vs posición)
#Se crean dos lista para guardas los valores de las abscisas y las coordenadas
xc=[]
yc=[]
#Se asume que la amplitud máxima (intensidad máxima) se encuentra en el centro del arreglo de circulo_nuevo
maxa=d2[512][512]
#Para hallar los valores de las coordenadas se elige toda la fila central del arreglo (circulo_nuevo[512,:]) se encuentra su norma
#y se eleva al cuadrado, se divide entre el la norma del valor máximo de intensidad (maxa) al cuadrado (para normalizar) 
yc= (abs(d2[512,:])**2)/abs(maxa)**2
#Se toman los valores para las abscisas de esta manera para poder comparar estos resultados con la teoría facilmente.
xc= 2*np.pi*0.01*np.linspace(-512,512,1024)
```

```python
#SE grafica la intensidad de la interferencia con respecto a la posición
#Se grafica tambiénla doble rendija junto con la transformada centrada en cero en un cuadrado específico
figure(figsize=(14,9))
subplot2grid((3,2),(0,0),colspan=2,rowspan=1)
ylabel(r"$I/I_0$")
xlabel(r"$2 \pi R j/L$")
plot(xc,yc,'o')
xlim(0,10)
ylim(0,1)
subplot2grid((3,2),(1,0),rowspan=2)
imshow(d1,cmap=cm.gray)
subplot2grid((3,2),(1,1),rowspan=2)
title(u'Patrón de interferencia de una apertura circular (Patrón de Airy)\n log resaltado')
final=(log((abs(d2))**2+1))
imshow(final)
tight_layout()
show()
```

```python
# Puntos escogidos para hacer la interpolación, se toman la mitad de los puntos dados de manera uniforme
xp=[]
yp=[]
for i in range(512):
    xp.append(xc[i*2])
    yp.append(yc[i*2])
#Se hace una interpolación cúbica con los puntos escogidos
interpol_cubic = interpolate.interp1d(xp,yp,kind='cubic')
#Se crea una lista con los intervalos entre los cuales están máximos y mínimos esperados.
inter=[-1,3,4,6,8,9,11,12,13]
#Se crea una lista minimos para guardar los mínimos encontrados, y mivalor para guardar el valor del mínimo correspondiente
minimos=[]
mivalor=[]
#Se utiliza el comando fminbound para encontrar el punto en x donde se encuentra un mínimo, a este comando se le da como parámetro
#la función , y el rango donde es posible encontrar el mínimo (se uso la lista inter que tiene estos rangos)
for i in range (3):
    minimos.append(fminbound(interpol_cubic,inter[i*2+1],inter[i*2+3]))
    #Se usa la interpolacion para encontrar el valor del mínimo
    mivalor.append(float(interpol_cubic(minimos[i])))
#Debido a que no hay una función equivalente a fminbound para hallar máximos lo que se hace es reflejar la función sobre el eje x
#Esta nueva función se determina así: se toman los puntos anteriores escogidos para y (yp) y se reflejan sobre el eje x (se multi-
#plican por -1 y se guardan en una nueva lista ypm
ypm=[]
for i in range(len(yp)):
    ypm.append(-(yp[i]))
#Se encuentra una interpolación cúbica para estos puntos
interpol_cubic_max = interpolate.interp1d(xp,ypm,kind='cubic')
#Se crean dos lista, maximos para guardar el punto donde se ubica el máximo y mavalor para guardar el valor de éste
maximos=[]
mavalor=[]
for i in range (4):              
    maximos.append(fminbound(interpol_cubic_max,inter[i*2],inter[i*2+2]))
    mavalor.append(float(interpol_cubic(maximos[i])))
```

```python
#En una sola lista se guardan máximo y mínimos y en otra, se guardan los valores correspondientes.
puntosx=[]
puntosy=[]
for i in range (len(minimos)):
    puntosx.append(maximos[i])
    puntosy.append(mavalor[i])
    puntosx.append(minimos[i])
    puntosy.append(mivalor[i])
puntosx.append(maximos[3])
puntosy.append(mavalor[3])
```

```python
#Puntos teóricos para hacer la comparación
xt= [0,3.833,5.136,7.016,8.417,10.174,11.620]
yt= [1,0,0.0175,0,0.0042,0,0.0016]

for i in range(len(xt)):
    if (i%2==0):
        print "Hay un máximo en " + str(puntosx[i]) + " y su valor es " + str(puntosy[i])
        print "Según la teoría el máximo se encuentra en " + str(xt[i]) + " y su valor es " + str(yt[i]) + "\n"
        
    if (i%2!=0):
        print "Hay un mínimo en " + str(puntosx[i]) + " y su valor es " + str(puntosy[i])
        print "Según la teoría el mínimo se encuentra en " + str(xt[i]) + " y su valor es " + str(yt[i]) + "\n"
```
