#a)

```python
#Se crea un arreglo bidimensional de dimensiones 1024x1024 inicializado en cero por practicidad
a=[[0 for i in range(1024)] for i in range(1024)]
#Se crean variables para facilitar la creacion de unos en el centro
lado=32 #Lado del cuadrado central de unos
liminf=len(a)/2-lado/2-1 #Límite inferior
limsup=len(a)/2+lado/2-1 #Límite superior
```

```python
#Se reemplazan los valores del cuadrado cental con unos
for i in range(liminf,limsup):
    for j in range(liminf,limsup):
        a[i][j]=1
#Se grafican en escala de grises: Negro es 0 y blanco es 1.
figure(figsize=(5,5))
imshow(a,cmap=cm.gray)
show()
```

#b)

```python
figure(figsize=(10,6))
#Se realiza la transformada de fourier en 2-D para el arreglo crado anteriormente
trans=fft2(a)
subplot(1,2,1)
#Se muestra la transformada directa y...
imshow(abs(trans))
title('Transformada de Fourier')
trans=fftshift(trans)
subplot(1,2,2)
#Se muestra la transformada centrada en cero
title('Transformada centrada en cero')
imshow(abs(trans))
show()
```

```python
#Se definen nuevas variables para el cuadrado a tratar
lado2=200
liminf=len(a)/2-lado2/2-1
limsup=len(a)/2+lado2/2-1
```

```python
#Se crea un nuevo arreglo bidimensional de tipo complejo con las dimensiones del nuevo cuadrado
b=zeros((lado2,lado2),complex)
#Se llena el nuevo arreglo con los valores deseados de la transformada centrada en cero
for i in range(lado2):
    for j in range(lado2):
        b[i][j]=trans[i+liminf][j+liminf]
imshow(abs(b))
```

```python
#Se crea el arreglo del logaritmo del modulo al cuadrado
final=(log((abs(b))**2))
#Y se grafica junto con la apertura correspondiente
figure(figsize=(15,9))
subplot(1,2,1)
imshow(a,cmap=cm.gray)
subplot(1,2,2)
title(u'Patrón de interferencia de una apertura cuadrada \n (log resaltado)')
imshow(final)
```

```python
#Se crea un arreglo bidimensional de dimensiones 1024x1024 inicializado en cero por practicidad
c1=[[0 for i in range(1024)] for i in range(1024)]
#Se crean variables para facilitar la creación de las rendijas
alto=200
ancho=20
ej1=460
ej2=564
liminf=len(c1)/2-alto/2-1
limsup=len(c1)/2+alto/2-1
limi1=ej1-ancho/2-1
limd1=ej1+ancho/2-1
limi2=ej2-ancho/2-1
limd2=ej2+ancho/2-1
#Se crea el arreglo con la doble rendija representada por unos en una matriz de ceros
for i in range(liminf,limsup):
    for j in range(limi1,limd1):
        c1[i][j]=1
    for k in range(limi2,limd2):
        c1[i][k]=1
#Segrafica a blanco y negro
imshow(c1,cmap=cm.gray)
```

```python
#Se hace la transformada de fourier para la doble rendija
trans=fft2(c1)
figure(figsize=(10,6))
subplot(1,2,1)
title('Transformada de Fourier')
imshow(abs(trans))
#Y se grafica junto con la transformada centrada en cero
trans=fftshift(trans)
subplot(1,2,2)
title('Transformada centrada en cero')
imshow(abs(trans))
```

```python
#Se crean las variables necesarias para seleccionar un cuadrado especifico del arreglo
lado=128
c2=zeros((lado,lado),complex)
liminf=len(c1)/2-lado/2-1
limsup=len(c1)/2+lado/2-1
#Se llena el nuevo arreglo con los valores deseados de la transformada centrada en cero
for i in range(lado):
    for j in range(lado):
        c2[i][j]=trans[i+liminf][j+liminf]
#Se grafica la doble rendija junto con la transformada centrada en cero en un cuadrado específico
figure(figsize=(15,9))
subplot(1,2,1)
imshow(c1,cmap=cm.gray)
subplot(1,2,2)
title(u'Patrón de difracción de una doble rendija')
imshow(abs(c2))
show()
```
