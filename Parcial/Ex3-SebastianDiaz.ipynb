```python
#Número de puntos
N=1000
M=50
```

##1. Coordenadas de puntos en un cuadrado unitario

```python
#Función que recibe un N y arroja un array de dos columnas con los puntos x y y de los N puntos
def puntos(N):
    pun=zeros((N,2))
    pun[:,0]=random.random(N)
    pun[:,1]=random.random(N)
    return pun
```

```python
#Prueba
pun=puntos(N)
scatter(pun[:,0],pun[:,1])
ylim(0,1)
xlim(0,1)
print puntos(N)
```

##2. Distancias no redundantes entre puntos

```python
def distancias(pun):
    distx=[]
    disty=[]
    for i in range(N):
        for j in range(N-1,i,-1):
            if(i!=j):
                distx.append(abs(pun[i,0]-pun[j,0]))
                disty.append(abs(pun[i,1]-pun[j,1]))
    distx=array(distx)
    disty=array(disty)
    dist=sqrt(distx**2+disty**2)
    return dist
```

```python
min(distancias(pun))
```

##3.Pseudocódigo
+ Una vez implementadas las dos funciones anteriores, lo primero que se debe hacer es hallar el mínimo valor del arreglo de las distancias.<BR>
+ Para eso se usa el comando `min()` que le entra por parametro el arreglo.<BR>
+ Este comando se debe usar en un `for` que realice un número **M** de ciclos sobre un mismo **N**, para, de esta manera, correr **M** veces los numeros aleatorios y encontrar la media del valor mínimo.
+ Una vez encontrada la media del valor mínimo para un **N** determinado, se implementa lo anterior dentro de un `for` más grande que itere sobre todos los **N**s deseados.

```python
promedios=[]
x=[2,5,8,20,50,80,200]
for i in (x):
    suma=0.
    for __ in range(M):
        punt=puntos(N)
        dist1=distancias(punt)
        suma+=min(dist1)
    promedios.append(suma/M)
promedios=array(promedios)
```

```python
figure()
title(u'Distancias mínimas promedio')
ylabel(r'$\vec{d}_m$')
loglog(x,promedios,'bo',basex=10)
```
