```python
%pylab inline
from scipy.fftpack import ifft, fft, fftfreq
from scipy.io.wavfile import write
from scipy import interpolate
```

#Punto 1

```python
#Se importan los datos
data=genfromtxt("ring.dft", delimiter="\n", dtype=complex)
inv=ifft(data)
f=22050
```

```python
print inv[8]
plot(real(inv))
```

```python
song=int16(real(inv))
```

```python
type (song[9])
```

```python
write("cancion.wav",f,song)
```

#Punto 2a

```python
y=real(inv)
N=len(y)
dt=1./f
x=linspace(0,N*dt,N)
```

```python
plot(x,y)
```

```python
mario= interpolate.interp1d(x,y,kind='slinear')
```

```python
newx=linspace(0,N*dt,2*N)
newy=mario(newx)
```

```python
plot(newx,mario(newx))
```

#Punto 2b

```python
fadeout=linspace(1,0,2*N)
plot(fadeout)
```

```python
fadeout_mario=newy*fadeout
plot(fadeout_mario)
```

```python
newsong=int16(fadeout_mario)
type (newsong[87])
```

```python
write("newsong.wav",f,newsong)
```
