```python
%pylab inline
from IPython.display import HTML
from scipy.signal import *
from scipy.fftpack import *
```

```python
ECG=genfromtxt('100.csv',delimiter=",")
t=ECG[:,0]
V=ECG[:,1]
prueba=V#[0:300000]
tn=t#[0:300000]
```

```python
for i in range(len(prueba)):
    if (prueba[i]<0):
        prueba[i]=0
arg=arange(1,300,300)
fp=find_peaks_cwt(prueba,arg,min_length=1)
```

```python
ar=[]
for i in range(len(fp)-1):
    if((fp[i+1]-fp[i])>150):
        ar.append(fp[i])
    if(i==len(fp)-2):
        ar.append(fp[i+1])
```

```python
plot(prueba)
for i in range(len(ar)):
    axvline(ar[i],color='r')
xlim(0,1000)
```

```python
tiempo=[]
tiempo.append(tn[ar[0]])
for i in range(1,len(ar)-1):
    tiempo.append(tn[ar[i+1]]-tn[ar[i]])
x=linspace(0,len(tiempo),len(tiempo))
plot(x,tiempo)
std(tiempo)
```

```python
posiciones=[]
print len(tiempo), len(ar)
for i in range(len(ar)-1):
    if(tiempo[i]<0.6):
        posiciones.append(ar[i+1])
print (posiciones[0])
```
