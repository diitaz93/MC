```python
%pylab inline
from IPython.display import HTML
from scipy.signal import *
from scipy.fftpack import *
```

```python
ECG=genfromtxt('noise.csv',delimiter=",")
t=ECG[:,0]
V=ECG[:,1]
fs=360 #Hz
dt=1./fs
N=len(V)
```

```python
prueba=V[5000:8000]
```

##Intento 1 Fourier

```python
prueba=V[5000:8000]
y=fft(prueba)
f=fftfreq(len(prueba),dt)
figure(figsize=(18,20))

#Transformada original
subplot(6,2,1)
plot(f,abs(y))
ylim(0,600)
xlim(-50,50)
#Señal original
subplot(6,2,2)
plot(prueba)
xlim(1000,1500)
title(u'Señal original')
#Transformada filtro 40
subplot(6,2,3)
y[abs(f)>40]=(10**-10)*angle(y[abs(f)>40])
plot(f,abs(y))
ylim(0,600)
xlim(-50,50)
#Señal filtro 40
subplot(6,2,4)
plot(ifft(y))
xlim(1000,1500)
title(u'Señal filtro 40')
#Transformada filtro 30
subplot(6,2,5)
y[abs(f)>30]=(10**-10)*angle(y[abs(f)>30])
plot(f,abs(y))
ylim(0,600)
xlim(-50,50)
#Señal filtro 30
subplot(6,2,6)
plot(ifft(y))
xlim(1000,1500)
title(u'Señal filtro 30')
#Transformada filtro 20
subplot(6,2,7)
y[abs(f)>20]=(10**-10)*angle(y[abs(f)>20])
plot(f,abs(y))
ylim(0,600)
xlim(-50,50)
#Señal filtro 20
subplot(6,2,8)
plot(ifft(y))
xlim(1000,1500)
title(u'Señal filtro 20')
#Transformada filtro 10
subplot(6,2,9)
y[abs(f)>10]=(10**-10)*angle(y[abs(f)>10])
plot(f,abs(y))
ylim(0,600)
xlim(-50,50)
#Señal filtro 10
subplot(6,2,10)
plot(ifft(y))
xlim(1000,1500)
title(u'Señal filtro 10')
#Transformada filtro 5
subplot(6,2,11)
y[abs(f)>5]=(10**-10)*angle(y[abs(f)>5])
plot(f,abs(y))
ylim(0,600)
xlim(-50,50)
#Señal filtro 5
subplot(6,2,12)
plot(ifft(y))
xlim(1000,1500)
title(u'Señal filtro 5')
#xlim(-50,50)
#ylim(0,1000)
show()
```

```python
#Filtro total fc=17 Hz
newy=fft(V)
newf=f=fftfreq(len(V),dt)
newy[abs(newf)>17]=(10**-10)*angle(newy[abs(newf)>17])
filt=ifft(newy)
figure(figsize=(23,10))
subplot(2,1,1)
plot(V)
xlim(3000,8000)
subplot(2,1,2)
plot(filt)
xlim(3000,8000)
show()
```

```python
ar=arange(1,300)
pks=find_peaks_cwt(filt[2000:8000],ar)
figure(figsize=(20,5))
plot(t[2000:8000],filt[2000:8000])
plot(t[2000:8000][pks],filt[pks],'ro')
```

```python
r=argrelmax(filt[3000:8000],order=160)
picos=[]
for i in range(len(r[0])):
    picos.append(r[0][i])
picos=array(picos)
figure(figsize=(20,5))
plot(filt[3000:8000])
plot(picos,filt[picos],'ro')
```
