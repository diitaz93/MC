```python
%pylab inline
from IPython.display import HTML
from scipy.signal import *
from scipy.fftpack import *
```

##Análisis de datos para el archivo 100.csv

```python
#Se importan los datos del archivo 100.csv
ECG100=genfromtxt('100.csv',delimiter=",")
t100=ECG100[:,0]
V100=ECG100[:,1]
#Gráfica de una parte de los datos
figure(figsize=(13,5))
plot(V100)
xlim(0,6000)
title("Voltaje")
ylabel(r'V (mV)')
xlabel(u"Número del dato")
```

```python
#Se desea encontrar la frecuencia cardiaca de la persona, para esto primero se va a hallar el período de las
#pulsaciones, la señal se limpia, se encuentran los picos máximos y usando esto el tiempo en el cuál ocurre
#cada pulso, luego el período es la diferencia entre dos tiempos consecutivos.
Vlim100=array(V100)
for i in range(len(Vlim100)):
    # Todos los voltajes negativos se igualan a cero, de esta manera solo sobrevive el complejo QRS (el pulso mas alto)
    if (Vlim100[i]<0):
        Vlim100[i]=0
figure(figsize=(13,5))
ylabel(r'V (mV)')
xlabel(u"Número del dato")
title(u"Señal del voltaje modificada")
plot(Vlim100)
xlim(0,6000)
show()
```

```python
#Ahora se desea hallar dónde se encuentran los máximos
arg100=arange(1,300,300)
#La función find peaks cwt recibe como principal parámetro el arreglo y devuelve un arreglo que contiene el número
#del dato donde se encuentra cada pulso (máximo local).
fp100=find_peaks_cwt(Vlim100,arg100,min_length=1)
```

```python
#Debido a que la señal del voltaje tiene algunas irregularidades, la función find peaks cuenta más picos de los que
#hay en realidad, cuenta algunos máximos locales como picos. La manera en que esto se soluciono fue analizando que
#tan lejos se encontraban los datos donde habian pulsos entre si (si habian mas de 150 datos entre ellos se conservaba
#el valor de ese dato).
ar100=[]
for i in range(len(fp100)-1):
    if((fp100[i+1]-fp100[i])>150):
        ar100.append(fp100[i])
    if(i==len(fp100)-2):
        ar100.append(fp100[i+1])
```

```python
figure(figsize=(13,5))
title(u"Puntos donde se detectaron pulsos")
plot(V100)
#plot(V100)
xlim(0,6000)
ylabel(r'V (mV)')
xlabel(u"Número del dato")
for i in range(len(ar100)):
    axvline(ar100[i],color='r')
```

```python
#En el arreglo tiempo100 se guarda el tiempo que transcurre entre dos pulsos consecutivos en segundos
tiempo100=[]
#En el arreglo tiem100 se guarda el valor del tiempo correspondiente a cada pulso
tiem100=[]
for i in range(len(ar100)-1):
    tiempo100.append(t100[ar100[i+1]]-t100[ar100[i]])
    tiem100.append(ar100[i+1])
frecuencia100=[]
#Promedio de la frecuencia cardíaca:
fre100=0
for i in range(len(tiempo100)):
    #La frecuencia es el inverso del período, para que sus unidades sean por minuto se multiplica por 60.
    frecuencia100.append(1/tiempo100[i] *60)
    fre100+= frecuencia100[i]
fre100=fre100/len(frecuencia100)
figure(figsize=(13,5))
title(u"Frecuencia cardíaca")
ylabel("Frecuencia (ipm)")
xlabel(u"Tiempo (s)")
xlim(0,225000)
axhline(fre100,color='r',label="Frecuencia promedio = " + str(fre100))
legend()
plot(tiem100,frecuencia100)
```

```python
#Se desea determinar que tanto varia el valor de la frecuencia del individuo a lo largo de tiempo, para esto se 
#usa la función que calcula la desviación estándar de un arreglo.
print "La desviación estándar de los datos de frecuencia cardiaca es de ", std(frecuencia100)
```

```python
#Ahora, se desea calcular el número de aritmias que el paciente presentó (número de frecuencias que tienen un valor 
#muy por debajo del promedio (68). Además, se quiere determinar el número del dato donde ocurrió una.
#En el arreglo posiciones100 se va a guardar el número del dato donde ocurre una aritmia.
posiciones100=[]
for i in range(len(ar100)-1):
    if(frecuencia100[i]<68):
        posiciones100.append(ar100[i+1])
```

```python
figure(figsize=(18,8))
plot(t100,V100)
for i in range(len(posiciones100)):
    axvline(t100[posiciones100[i]],color='r')
xlim(t100[60000],t100[100000])
xlabel('Tiempo(s)')
ylabel('Voltaje(mV)')
print "Hay", len(posiciones100), "aritmias en total, estas se encuentran en los tiempos:", t100[posiciones100]
```

```python
#A simple vista puede observarse una aritmia en este punto, el pulso se demora mas de lo normal
figure(figsize=(10,5))
plot(t100,V100)
axvline(t100[posiciones100[3]],color='r')
xlim(t100[90000],t100[110000])
xlabel('Tiempo(s)')
ylabel('Voltaje(mV)')
```

Se determino que la frecuencia cardíaca del individuo es de 76.3 pulsaciones por minuto, la frecuencia cardíaca
promedio de una persona es de 70 latidos por minuto, un valor que esta un poco por debajo. Pero, por lo general
la frecuencia normal en reposo oscila entre 50 y 100 latidos por minuto. La desviación estándar de los datos indica
que tanto cambia la frecuencia cardiaca de la persona.

##Análisis de datos para el archivo 119.csv

```python
#Se importan los datos del archivo 119.csv
ECG9=genfromtxt('119.csv',delimiter=",")
t9=ECG9[:,0]
V9=ECG9[:,1]
#Gráfica de una parte de los datos
figure(figsize=(13,5))
plot(V9)
xlim(0,6000)
title("Voltaje")
ylabel(r'V (mV)')
xlabel(u"Número del dato")
```

```python
#Para encontrar la frecuencia cardiaca del paciente se realiza el mismo procedimiento anterior
#Tratamiento de la señal
Vlim9=array(V9)
for i in range(len(Vlim9)):
    #Los potenciales negativos se igualan a cero
    if (Vlim9[i]<0):
        Vlim9[i]=0
arg9f=arange(1,300,300)
#Se encuentran los puntos donde se ubican máximos locales
fp9f=find_peaks_cwt(Vlim9,arg9f,min_length=1)
#Se eliminan los datos que esten muy cerca entre sí (para contar cada máximo una sola vez)
ar9f=[]
for i in range(len(fp9f)-1):
    if((fp9f[i+1]-fp9f[i])>150):
        ar9f.append(fp9f[i])
    if(i==len(fp9f)-2):
        ar9f.append(fp9f[i+1])
figure(figsize=(13,5))
title(u"Puntos donde se detectaron pulsos")
plot(V9)
xlim(0,6000)
ylabel(r'V (mV)')
xlabel(u"Número del dato")
for i in range(len(ar9f)):
    axvline(ar9f[i],color='r')
```

```python
#En el arreglo tiempo9 se guarda el tiempo que transcurre entre dos pulsos consecutivos en segundos
tiempo9=[]
#En el arreglo tiem9 se guarda el valor del tiempo correspondiente a cada pulso
tiem9=[]
for i in range(len(ar9f)-1):
    tiempo9.append(t9[ar9f[i+1]]-t9[ar9f[i]])
    tiem9.append(ar9f[i+1])
frecuencia9=[]
#Promedio de la frecuencia cardíaca:
fre9=0
for i in range(len(tiempo9)):
    #La frecuencia es el inverso del período, para que sus unidades sean por minuto se multiplica por 60.
    frecuencia9.append(1/tiempo9[i] *60)
    fre9 += frecuencia9[i]
fre9=fre9/len(frecuencia9)
figure(figsize=(13,5))
title(u"Frecuencia cardíaca")
ylabel("Frecuencia (ipm)")
xlabel(u"Tiempo (s)")
xlim(0,225000)
axhline(fre9,color='r',label="Frecuencia promedio = " + str(fre9))
legend()
plot(tiem9,frecuencia9)
```

```python
#Se desea determinar que tanto varia el valor de la frecuencia del individuo a lo largo de tiempo, para esto se 
#usa la función que calcula la desviación estándar de un arreglo.
print "La desviación estándar de los datos de frecuencia cardiaca es de", std(frecuencia9)
```

```python
#Se calcula el número de aritmias que el paciente presenta usando el mismo procedimiento anterior
posiciones9=[]
for i in range(len(ar9f)-1):
    if(frecuencia9[i]<60):
        posiciones9.append(ar9f[i+1])
figure(figsize=(18,8))
plot(t9,V9) 
for i in range(len(posiciones9)):
    axvline(t9[posiciones9[i]],color='r')
xlim(t9[0],t9[100000])
print "Hay", len(posiciones9), "aritmias en total, estas se encuentran en los datos", posiciones9
```

```python
#El siguiente objetivo es calcular la cantidad de fibrilaciones que presenta el paciente
#Se manipula la onda para facilitar la búsqueda de las fibrilaciones(Vs)
Vlim29=array(V9)
#Dado que el valor de los Vs es negativo, todos los potenciales positivos se igualan a cero, al igual que los
#valores negativos pequeños (>-2). Los puntos restantes se reflejan sobre el eje x para poder usar la función find peaks 
#para encontrar máximos.
for i in range(len(Vlim29)):
    #
    if (Vlim29[i]>0):
        Vlim29[i]=0
    elif (Vlim29[i]>-2):
        Vlim29[i]=0
    else:
        Vlim29[i]=- Vlim29[i]
arg9=arange(1,300,300)
#Se encuentran los máximos de la nueva onda
fp9=find_peaks_cwt(Vlim29,arg9,min_length=1)
```

```python
ar9=[]
#Se guarda sólo un punto por cada V 
for i in range(len(fp9)-1):
    if((fp9[i+1]-fp9[i])>150):
        ar9.append(fp9[i])
    if(i==len(fp9)-2):
        ar9.append(fp9[i+1])
print "La cantidad de fibrilaciones que presenta el paciente son", len(ar9), "y la gráfica muestra en que puntos se presentan."
figure(figsize(15,8))
plot(t9,V9)
for i in range(len(ar9)):
    axvline(t9[ar9[i]],color='r')
title("Voltaje")
ylabel(r'V (mV)')
xlabel(u"Tiempo(s)")
```
