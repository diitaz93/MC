```python
Nt=2000 #Numero de datos de tiempo
Nl=50 #Numero de datos de posición
L=10. # Lado del cubo
tmin=0
tmax=.3
xmin=-L/2
xmax=L/2
ymin=-L/2
ymax=L/2
dt=(tmax-tmin)/Nt
dx=(xmax-xmin)/Nl
v=425. # m/s
xcoords=np.linspace(xmin,xmax,Nl)
ycoords=np.linspace(xmin,xmax,Nl)
xmesh,ymesh=np.meshgrid(xcoords,ycoords)
tcoords=np.linspace(tmin,tmax,Nt)
r=v*dt/dx
print "r=",r #Debe ser menor a 0.5
```

```python
onda=zeros((Nt,Nl,Nl))
lados=zeros((Nt,Nl,Nl))
tapas=zeros((Nt,Nl,Nl))
onda[0]=np.exp(-xmesh**2)*np.exp(-ymesh**2)
onda[1]=onda[0]+r**2/2.*(np.roll(onda[0],1,axis=0)+np.roll(onda[0],-1,axis=0)+np.roll(onda[0],1,axis=1)+np.roll(onda[0],-1,axis=1)-4*onda[0])
```

```python
figure(figsize=(12,5))
subplot(1,2,1)
imshow(onda[0],cmap='gist_heat')
subplot(1,2,2)
pcolormesh(xmesh,ymesh,onda[0],cmap='gist_heat')
xlim(-5,5)
ylim(-5,5)
colorbar()
show()
```

```python
LO=zeros((Nt,2*Nl,Nl))
for i in range(2*Nl):
    for j in range(Nl):
        if (i<=Nl-1):
            LO[0,i,j]=lados[0,i,j]
        else:
            LO[0,i,j]=onda[0,i-Nl,j] 
imshow(LO[0],cmap='gist_heat')
LO[1]=LO[0]+r**2/2.*(roll(LO[0],1,axis=0)+roll(LO[0],-1,axis=0)+np.roll(onda[0],1,axis=1)+np.roll(onda[0],-1,axis=1)-4*onda[0])
```

```python
TO=zeros((Nt,Nl,2*Nl))
for i in range(Nl):
    for j in range(2*Nl):
        if (j<=Nl-1):
            TO[0,i,j]=tapas[0,i,j]
        else:
            TO[0,i,j]=onda[0,i,j-Nl] 
imshow(TO[0],cmap='gist_heat')
```

```python
for i in range(2,Nt):
    ponda=onda[i-1]
    pponda=onda[i-2]
    pLO=LO[i-1]
    pTO=TO[i-1]
    plf1=roll(pLO,1,axis=1)
    plb1=roll(pLO,-1,axis=1)
    ptf2=roll(pTO,1,axis=2)
    ptb2=roll(pTO,-1,axis=2)
    onda[i]=2*(1.-2.*r**2)*ponda-pponda+r**2*(plf1[i,Nl:-1]+plb1[i,Nl:-1]+ptf2[i,:,Nl:-1]+ptb2[i,:,Nl:-1])
   
```
