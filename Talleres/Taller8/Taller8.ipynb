```python
Nt=2000 #Numero de datos de tiempo
Nl=50 #Numero de datos de posición
L=10. # Lado del cubo
tmin=0
tmax=.3
xmin=-L/2
xmax=L/2
ymin=-L/2
ymax=L/2
dt=(tmax-tmin)/Nt
dx=(xmax-xmin)/Nl
v=425. # m/s
xcoords=np.linspace(xmin,xmax,Nl)
ycoords=np.linspace(xmin,xmax,Nl)
xmesh,ymesh=np.meshgrid(xcoords,ycoords)
tcoords=np.linspace(tmin,tmax,Nt)
r=v*dt/dx
print "r=",r #Debe ser menor a 0.5
```

```python
onda=zeros((Nt,Nl,Nl))
lados=zeros((Nt,Nl,Nl))
tapas=zeros((Nt,Nl,Nl))
onda[0]=np.exp(-xmesh**2)*np.exp(-ymesh**2)
#onda[1]=onda[0]+r**2/2.*\
       # (roll(onda[0],1,axis=0)+roll(onda[0],-1,axis=0)+roll(onda[0],1,axis=1)+roll(onda[0],-1,axis=1)\
        #-4*onda[0])
```

```python
figure(figsize=(12,5))
subplot(1,2,1)
imshow(onda[0],cmap='gist_heat')
subplot(1,2,2)
pcolormesh(xmesh,ymesh,onda[0],cmap='gist_heat')
xlim(-5,5)
ylim(-5,5)
colorbar()
show()
```

```python
LO=zeros((Nt,2*Nl,2*Nl))
LO[0,0:Nl,0:Nl]=onda[0]
imshow(LO[0],cmap='gist_heat')
```

```python
LO[1]=LO[0]+r**2/2.*(roll(LO[0],1,axis=0)+roll(LO[0],-1,axis=0)+roll(LO[0],1,axis=1)+roll(LO[0],-1,axis=1)-4*LO[0])
```

```python
for i in range(2,Nt):
    pLO=LO[i-1]
    ppLO=LO[i-2]
    plf0=roll(pLO,1,axis=0)
    plb0=roll(pLO,-1,axis=0)
    plf1=roll(pLO,1,axis=1)
    plb1=roll(pLO,-1,axis=1)
    LO[i]=2*(1.-2.*r**2)*pLO-ppLO+r**2*(plf0+plb0+plf1+plb1)
    onda[i]=LO[i,0:Nl,0:Nl]
    lados[i]=LO[i,0:Nl,Nl:]
    tapas[i]=LO[i,Nl:,0:Nl] 
    LO[i,Nl,Nl:]=LO[i,0,Nl:]
    LO[i,Nl:,Nl]=LO[i,Nl:,0]
```

```python
figure(figsize=(14,14))
subplot(2,2,1)
imshow(LO[-1],cmap='gist_heat')
subplot(2,2,2)
imshow(onda[-1],cmap='gist_heat')
subplot(2,2,3)
imshow(lados[-1],cmap='gist_heat')
subplot(2,2,4)
imshow(tapas[-1],cmap='gist_heat')
show()
```
