```python
%pylab inline
from scipy.integrate import odeint
from scipy.integrate import ode
from mpl_toolkits.mplot3d import axes3d
```

```python
#Valores de los coeficientes de las ecuaciones diferenciales
a=0.1
b=a
c=14.
#Vector tiempo
t=linspace(0,600,50000)
```

```python
#Se define la función con las ecuaciones diferenciales
def func(Y,t):
    return array([-Y[1]-Y[2],Y[0]+a*Y[1],b+Y[2]*(Y[0]-c)])
```

```python
#Se crea una matriz solución con odeint 
funsol=odeint(func,[1.,0.,1.],t)
x=funsol[:,0]
y=funsol[:,1]
z=funsol[:,2]
```

```python
#Gráfica 2-D
figure(figsize=(10,10))
plot(x,y)
show()
```

```python
#Gráfica 3-D
figure(figsize=(15,12))
ax=gca(projection='3d')
ax.plot(x,y,z)
show()
```

```python
cs=array([4,6,8.5,8.7,9,12,12.6,13,18])
figure(figsize=(12,12))

for i in range(len(cs)):
    c=cs[i]
    cw=str(c)
    cf=str('c='+cw)
    funsol=odeint(func,[1.,0.,1.],t)
    x=funsol[5000:,0]
    y=funsol[5000:,1]
    subplot(3,3,i+1)
    plot(x,y,lw=0.3)
    title(cf)   
show()
```

```python
l=linspace(1,45,200)
figure(figsize=(15,10))
xlim(0,45)
xlabel('c')
ylim(0,60)
#Este ciclo varia el valor de c
for i in range(len(l)):
    py=[]
    px=[]
    c=l[i]
    #Se resuelve el sistema de ecuaciones diferenciales para cada valor de c
    funsol=odeint(func,[1.,0.,1.],t)
    #Se escogen los arreglos para que eliminen el transiente de la solución
    x=funsol[5000:,0]
    y=funsol[5000:,1]
    #Este ciclo recorre los arreglos x y y
    for j in range(len(y)):
        #Cuando el punto esta dentro de la seccion de Poincaré...
        if((y[j]>=0.) & (abs(x[j])<=1.)):
            #Se agregan sus coordenadas x y y a dos arreglos nuevos
            py.append(y[j])
            px.append(c)
    py=array(py)
    px=array(px)
    scatter(px,py,c='k',s=0.2)
    title('Rossler System: Bifurcation Diagram\n a=b=0.1')  
show()
```
