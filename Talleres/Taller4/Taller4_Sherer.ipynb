#Polinomios de Lagrange
Se quiere lograr la siguiente fórmula:
    
$\sum\limits_{j=0}^k y_j\prod\limits_{m=0}^k \frac{x-x_m}{x_j-x_m}$

```python
#Se hace la multiplicatoria mostrada arriba
def lagrangep(absc,j):
    pol=np.poly1d([0,1])
    k=len(absc)
    for m in range(k):
        if m!=j:
            pol*=1./(absc[j]-absc[m])*np.poly1d([1.,-absc[m]])
    return pol 
```

```python
#Se acopla la multiplicatoria declarada anteriormente con la sumatoria
def interlagr(absc,orde):
    poly=np.poly1d([0])
    for i in range(len(absc)):
        poly+=orde[i]*lagrangep(absc,i)
    return poly
```

```python
#Se crea un arreglo con puntos específicos y se evalua la función seno en ellas
x=[0,pi/2,pi,3*pi/2,2*pi]
y=sin(x)
#Se interpola con estos puntos
func=interlagr(x,y)
```

```python
#Se crea un arreglo más largo y se repite el paso anterior
x1=[0,pi/4, pi/2, 3*pi/4, pi, 5*pi/4, 3*pi/2, 7*pi/4, 2*pi]
y1=sin(x1)
func1=interlagr(x1,y1)
```

```python
#Se crea un arrgelo con más puntos, para evaluar los dos tipos de interpolaciones
newx=linspace(0,2*pi,100)
plot(x,y,'co')
plot(x1,y1,'b+')
plot(newx,func(newx))
plot(newx,func1(newx))
```

```python
#Se definen los arreglos
x=[-3,-2,-1,0,1,2,3]
delta=[0,0,0,1,0,0,0]
heaviside=[0,0,0,1,1,1,1]
```

```python
dinter=interlagr(x,delta)
hinter=interlagr(x,heaviside)
nx=linspace(-3,3,100)
```

```python
ndelta= interpolate.interp1d(x,delta,kind='zero')
stem(x,ndelta(x))
plot(nx,dinter(nx))
```

```python
nheav= interpolate.interp1d(x,heaviside,kind='zero')
stem(x,nheav(x))
plot(nx,hinter(nx))
```
