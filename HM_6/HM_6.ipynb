##Polinomios de Lagrange
Se quiere realizar una función que retorne simbólicamente la siguiente función:<BR>
$$L(x):=\sum_{j=0}^k f_i\prod_{\substack{0\le \le k \\ m\neq j}}\frac{x-x_m}{x_j-x_m}$$

```python
#Se definen las variables simbólicas a utilizar
pol ,x, PL=symbols('pol x PL')
```

```python
#Se define una función que retorne el polinomio de lagrange en símbolos dado un numero
def lagrange(N):
    #Se inicializa la variable pol en cero
    pol=0
    #El primer ciclo hace un recorrido por el numero entrado por parametro
    #que determinará cuántos términos habra en la sumatoria
    for j in range(N):
        #El indice del primer ciclo se convierte en string
        w=str(j)
        #Se crea una variable simbólica de las ordenadas con el subindice
        fj=sympify('f'+w)
        #Se inicializa la variable PL en 1. Esta es el termino de la multiplicatoria
        PL=1
        #Se crea una variable simbólica de las abscisas con el subindice
        xj=sympify('x'+w)
        for m in range(N):
            #Se convierte en string el indice del segundo ciclo
            s=str(m)
            #Se crea la variable simbólica con el indice. 
            xm=sympify('x'+s)
            #Evalua los subindices m y j, y solo arroja un polinomio cuando los subindices son diferentes
            if (m!=j):
                PL*=(x-xm)/(xj-xm)
        #Se multimplica la ordenada por el polinomio de la multiplicatoria
        pol+=fj*PL
    return pol
```

```python
#Función que retorna simbólicamente las ecuaciónes de adam bashford dado el orden
def adams(m):
    #Se crean las variables simbólicas con sus subindices cuando es necesario
    xm=symbols('x:%d'%m)
    h,x=symbols('h x')
    #Se crea un arreglo vacio
    arreglo=[]
    
    for i in range(m):
        #Se agregan filas al arreglo
        arreglo += [(xm[i],xm[0]-h*i)]
    pol=lagrange(m)
    #Integra el polinomio de lagrange
    pol= integrate(pol, (x,xm[0],xm[0]+h))
    #Se sustituye el termino xn por el de la columna 1 del arreglo
    pol= pol.subs(arreglo)
    return simplify(pol)
```

#Punto 2.a)
Se debe resolver la ecuación de *Poisson* gravitacional:
$$\frac{\partial ^2\Psi}{\partial x^2}+\frac{\partial ^2\Psi}{\partial y^2}+\frac{\partial ^2 \Psi}{\partial z^2}=-4G\pi\rho$$

```python
#Se declaran las constantes
G=6.67*10**-11
C=4.*pi*G
m=400.*(270.**3)
L=3000  
Nl=101 
xmin=-L/2
xmax= L/2
ymin=-L/2
ymax= L/2
zmin=-L/2
zmax= L/2
#Se crean los arreglos de las coordenadas
xcoords=np.linspace(xmin,xmax,Nl)
ycoords=np.linspace(ymin,ymax,Nl)
zcoords=np.linspace(zmin,zmax,Nl)
#Y las cuadriculas cara cada coordenada
xmesh,ymesh,zmesh=np.meshgrid(xcoords,ycoords,zcoords)
#Se define el stepsize
Dx=(xmax-xmin)/(Nl-1)
```

```python
#Se define un número de iteraciones para el método de la relajación
NumIter=10000
```

```python
#Masa puntual
#Se define el campo producido por una masa puntual en el centro
field=-G*m/sqrt(xmesh**2+ymesh**2+zmesh**2)
#Se evita el problema de -inf que ocurre al dividir por cero
field[50,50,50]=field[50,50,49]
```

```python
#Cubo de 270
density=zeros((Nl,Nl,Nl))
#Se distribuye la masa uniformemente en un cubo de 270m de lado
density[46:54,46:54,46:54]=400 
```

```python
#Se muestra el cubo de masa y el potencial de la masa puntual
figure(figsize=(10,4))
subplot(1,2,1)
pcolormesh(ymesh[:,50,:],zmesh[:,50,:],density[50,:,:],cmap='gist_heat')
subplot(1,2,2)
pcolormesh(ymesh[:,50,:],zmesh[:,50,:],field[50,:,:],cmap='gist_heat')
colorbar()
show()
```

**Metodo de la relajación:**<BR>
$\Phi_{i,j,k}=\frac{1}{6}(\Phi_{i+1,j,k}+\Phi_{i-1,j,k}+\Phi_{i,j+1,k}+\Phi_{i,j-1,k}+\Phi_{i,j,k+1}+\Phi_{i,j,k-1}-4G\pi\rho_{i,j})$

```python
#Método de la relajación
#Se toma la densidad como la distribución uniforme de masa en un cubo de lad 270 declarado anteriormente
scaled_density=C*Dx**2*density
#Iteración
for __ in range(NumIter):
    #Esta es la fórmula mostrada arriba
    field=1/6.*(roll(field,1,axis=0)+\
                roll(field,-1,axis=0)+\
                roll(field,1,axis=1)+\
                roll(field,-1,axis=1)+\
                roll(field,1,axis=2)+\
                roll(field,-1,axis=2)-scaled_density)
    #Condiciones de frontera despues de cada iteración
    field[0,:,:]=0.
    field[-1,:,:]=0.
    field[:,0,:]=0.
    field[:,-1,:]=0.
    field[:,:,0]=0.
    field[:,:,-1]=0.
```

```python
#Gráfica del potencial gravitacional en el plano y-z
plt.figure(figsize=(12,10))
plt.pcolormesh(ymesh[:,50,:],zmesh[:,50,:],10000*field[50,:,:],cmap='gist_heat')
title(r"$\Phi /cm^2 s^{-2}$",fontsize=20)
xlabel(r"$z/m$",fontsize=16)
ylabel(r"$y/m$",fontsize=16)
plt.colorbar()
CS=plt.contour(ymesh[:,50,:],zmesh[:,50,:],10000*field[50,:,:],[-1,-2,-4,-8,-16],colors='k')
plt.clabel(CS)
plt.show()
```

```python
nu=9
# se define la masa de cada sub cubo
mi=m/(9**3)
#En la lista posicion se guardan las posiciones del centro del cada subcubo, hay 9 centros de cubos en cada coorde
#nada.
posicion=[]
for i in range(nu):
    for j in range(nu):
        for k in range(nu):
            posicion.append([-120 + 30*i, -120 + 30*j,-120 +30*k])    
#En la lista zeta se guardan las posiciones en la coordenada z, donde el dx es 30 (el lado de los subcubos)
zeta=[]
for i in range(100):
    zeta.append(-1500+30*i)
zeta=array(zeta)
```

```python
# en la lista valores se guarda el valor del potencial correspondiente a cada valor de zeta (posicion en la coordena
#da zeta
valores=[]
for i in range(len(zeta)):
    #en suma se acumula la suma del potencial debido a cada subcubo
    suma=0
    for j in range(len(posicion)):
        #distancia entre la posicion del subcubo j y la posicion en el eje z
        distan= np.sqrt((posicion[j][0])**2 + (posicion[j][1])**2 + (posicion[j][2]- zeta[i])**2)
        #para evitar el problema de dividir en cero, a esta distancia se le asigna otro valor (el siguiente) 
        if (distan==0):
            distan= np.sqrt((posicion[j+1][0])**2 + (posicion[j+1][1])**2 + (posicion[j+1][2]- zeta[i])**2)
        #se suma el potencial debido al cubo j
        suma += (-G*mi)/distan
    # se hace el cambio de unidades
    valores.append(suma*10**4)
valores=array(valores)
```

```python
#Se obtiene una tajada del arreglo de potencial
ar=field[50,50,:]
#Se ajusta una constante para sumarle al potencial de masas puntuales para que sea acorde con el de masa continua 
eps=6*G*m
v=valores+eps
#Gráfica de el cálculo del potencial con la distribución continua de masas y con el equivalente de masas puntuales
figure(figsize=(15,10))
plt.scatter(xcoords,10000*ar,c='white',label='Poisson')
plt.plot(zeta,v,label=r'sum+ $\epsilon$')
xlim(-1500,1500)
ylim(-50,0)
title('Potencial Gravitacional')
ylabel(r'$\Phi/cm^2s^{-2}$',fontsize=16)
xlabel(r'$z/m$',fontsize=16)
legend(loc='lower right')
show()
```

##c) Se deriva el potencial gravitacional para obtener el campo gravitacional, por el método de diferencias centrales

```python
#Se define la longitud del paso
h=3000/(len(ar)-1)
#Se crea un arreglo del potencial adelantado una posición
arf=roll(ar,1)
#Y un arreglo del potencial atrasado una posición
arb=roll(ar,-1)
#Se crea el arreglo derivada por diferencias centrales
dar=(arf-arb)/(2.*h)
#Se eliminan los primeros y el últimos elementos de los arreglos ya que no tienen sentido físico
dar=dar[1:-1]
xc=xcoords[1:-1]
#Gráfica de la derivada
figure(figsize=(10,8))
scatter(xc,10**6*dar, c="white")
title('Campo Gravitacional')
xlabel(r'$g/m$',fontsize=16)
ylabel(r'$g/\mu ms-2}$',fontsize=16)
xlim(-1500,1500)
ylim(-20,20)
show()
```

```python
#Se hace una interpolacion de los puntos obtenidos
interpol_cubic = interpolate.interp1d(zeta[0:-1],dar,kind='cubic')
#Se define una funcion que recibe un arreglo Y, y su estructura es Y=[y,v](posicion y velocidad), y devuelve
# la derivada con respecto al tiempo (velocidad y aceleracion --> gravedad)
def f(Y):
    return np.array([Y[1],interpol_cubic(Y[0])])
#funcion runge kutta de grado 4
def one_Kutta4_step():
    global t
    global Deltat
    global Y
    K1 = f(Y)
    K2 = f(Y + Deltat*(1./2.*K1))
    K3 = f(Y + Deltat*(1./2.*K2))
    K4 = f(Y + Deltat*K3)
    Y+= Deltat * (1/6.*K1 + 1/3. * K2 + 1/3.*K3 + 1/6.*K4)
    t+= Deltat
#Condiciones iniciales:
Y= np.array([-500.-135.,0.])
numIter=400
tmax=30000
t=0.
Deltat=tmax/numIter
```

```python
#Se resuelve la ecuacion usando Runge-Kutta de orden 4
history=np.array([[t,Y[0],Y[1]]])
for i in range(numIter):
    one_Kutta4_step()
    history=np.append(history,[[t,Y[0],Y[1]]],axis=0)
#Gráfica de Posición vs Tiempo
figure(figsize=(8,6))
plt.plot(history[:,0],history[:,1])
title(u"Posición en z vs. tiempo")
xlabel("Tiempo(s)")
ylabel("Distancia en z (m)")
```
