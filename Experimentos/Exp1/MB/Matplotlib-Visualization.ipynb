#3. The Plot command

##The plot command would let you plot funtions and more.

```python
x = linspace(0, 2*pi, 100)
y = sin(x)
plot(x, y)
```

## You can change line colors using the flag $\verb+c+$

```python
plt.plot(x, y, c='k')
```

## You can change line width and line style using: $\verb+lw+\  \&\ $  $\ \ \ \verb+ls+$ 

```python
plt.plot(x, y, lw=3.0, ls='--',c='g')
```

## You can plot as many functions as you want in the same plot.

```python
plt.plot(x, y, lw=3.0, ls='--', c='b')
plt.plot(x, 2*y, lw=2.0, ls='-', c='k')
plt.plot(x, 3*y, lw=1.0, ls='-.', c='r')
plt.plot(x, 4*y,  "-o", c='r', ls='.-')
```

#4. Legends, captions , $\LaTeX \ \ \ \ \ \ \ \ \ \ \ \ \ \ $ style & Axys limits.

## Legends: You can add legends using label = 'your legend' and then calling the function plt.legend() 


```python
plot(x, y, lw=3.0, ls='--', c='b', label = 'sin(x)')
legend()
```

```python
## You can change the legend fontsize and the position
```

```python
plt.plot(x, y, lw=3.0, ls='--', c='b', label = 'sin(x)')
plt.legend(fontsize=20, loc='center')
```

## A powerfull tool of Matplotlib is that it allows you to write in $\LaTeX \ \ \ \ \ $ font style, for this you have to add the symbols $\$ $ before and 
## after the text and an r e.g:   $\verb+r'\$sin(x)\$'+$

```python
plt.plot(x, y, lw=3.0, ls='--', c='b', label = r'$\nabla$')
plt.legend(fontsize=20)

```

## Axis Labels and Limits

```python
plt.plot(x, y, lw=3.0, ls='-', c='b', label = r'$sin(x)$')
plt.legend(fontsize=20)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$sin(x)$', fontsize=20)
plt.title(r'$sin\ Function$', fontsize=30)
plt.xlim([0, 2*np.pi])
```

## Text command. first argument x coordinates, second argument y coordinate, third argument the text

```python
plt.plot(x, y, lw=3.0, ls='-', c='b', label = r'$sin(x)$')
plt.legend(fontsize=20)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$sin(x)$', fontsize=20)
plt.xlim([0, 2*np.pi])
plt.text(0.1, 1.5, "$\mathrm{You\ can\ also\ put\ text\ anywhere}$", fontsize=30, rotation='vertical')
```

## Axis font size! Using tick_params

```python
plt.plot(x, y, lw=3.0, ls='-', c='b', label = r'$sin(x)$')
plt.legend(fontsize=20)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$sin(x)$', fontsize=20)
plt.xlim([0, 2*np.pi])
plt.tick_params(axis="both", labelsize=20, colors="k")
```

#5. Figure Size and grid

```python
figure(figsize=(14, 10))
plt.plot(x, y, label='$sin(x)$')
plt.legend(fontsize=20)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$sin(x)$', fontsize=20)
plt.xlim([0, 2*np.pi])
plt.grid() 
```

# 6. Error bars (errorbar)

```python
figure(figsize=(7, 5))
plt.errorbar(x, y, yerr=0.2, label='$sin(x)$')
plt.legend(fontsize=20)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$sin(x)$', fontsize=20)
plt.xlim([0, 2*np.pi])
plt.tick_params(axis="both", labelsize=18, colors="k")
plt.grid() 
```

# 7. Log scale axis. 

```python
figure(figsize=(7, 5))
plt.errorbar(x*10E4, y*10E15, yerr=0.2, xerr=0.5, label='$sin(x)$')
plt.xscale("log")
plt.yscale("log")
plt.legend(loc='best', fontsize=20)
plt.xlabel(r'$x$', fontsize=30)
plt.ylabel(r'$sin(x)$', fontsize=30)
plt.xlim([10E2, 10E5])
plt.ylim([10E14, 10E15])
plt.tick_params(axis="both", labelsize=18, colors="k")
plt.grid() 
```

#8. Subplots

```python
plt.figure(figsize=(20, 16))
plt.subplots_adjust(hspace=0.00, wspace=0.00)
plt.subplot(221)
plt.scatter(x, y, c='r')
plt.subplot(222)
plt.scatter(x, y, c='g')
plt.subplot(223)
plt.scatter(x, y, c='y')
plt.subplot(224)
plt.scatter(x, y)
```

#9. Exporting the plots

```python
figure(figsize=(7, 5))
plt.errorbar(x, y, yerr=0.2, xerr=0.5, label='$sin(x)$')
plt.legend(fontsize=20)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$sin(x)$', fontsize=20)
plt.xlim([0, 2*np.pi])
plt.tick_params(axis="both", labelsize=18, colors="k")
plt.grid() 
plt.savefig("sinfunction.png", format='png', bbox_inches='tight', transparent=True)
plt.savefig("sinfunction.eps", format='eps', bbox_inches='tight', transparent=True)
plt.savefig("sinfunction.pdf", format='pdf', bbox_inches='tight', transparent=True)
```

# 10. Imshow

Now imagine you want to visualize the following matrix.  

```python
x = np.zeros([10,10])

for i in range(10):
    for j in range(10):
        if ((i-5)**2 + (j-5)**2 < 5):
            x[i][j] = 1
print x
```

```python
plt.imshow(x, interpolation="nearest")
```

you can try different interpolation modes to have different results.

```python
plt.imshow(x, interpolation="gaussian")
```

In most cases it would be important to know the values of the image. For this reason we need a colorbar.

```python
plt.imshow(x, interpolation="gaussian")
plt.colorbar()
```

Let's play with a non square matrix

```python
x = np.zeros([10,20])

for i in range(10):
    for j in range(20):
        if ((i)**2 + (j)**2 < 5):
            x[i][j] = 1

plt.imshow(x)
plt.colorbar(orientation = "horizontal")
```

```python
plt.imshow(x.T)
plt.colorbar()
```

# 11. 3d plots:

Some times you will need to show data in 3d space. To see usefull 3d plotting examples:
    
(http://matplotlib.org/gallery.html#mplot3d)

```python
# Taken from http://matplotlib.org/examples/mplot3d/wire3d_demo.html
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plt.show()

```
