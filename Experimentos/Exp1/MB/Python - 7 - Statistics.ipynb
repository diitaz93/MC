10.1 File I/O

Getting data into and out of files in Python feels a lot like using the low-level methods of C, but it has all the ease of Python layered on top.

For example, to open a file for writing use the builtin (and very C-like) open() function. But then write the contents of a list with a single call. The open() function returns an open file object and closing the file is done by calling the close() method of that object.

```python
colores = ['rojo\n','amarillo\n','azul\n']
efe = open('colores.txt','w')
efe.writelines(colores)
efe.close()
```

By default, the open() function returns a file open for reading. Individual lines can be read with the readline() method, which will return an empty string. Since the zero-length string has is not truthy, it makes a simple marker.

```python
efe=open('colores.txt','r')
print efe.readline(),
print efe.readline(),
print efe.readline(),
# Cuando no quedan líneas por leer se obtiene el string vacío
print efe.readline(),
efe.close()
```

Alternatively, all of the lines of the file can be read into a list with one method call and then iterated over from there.

```python
efe=open('colores.txt')
print efe.readlines()
efe.close()
```

However, for large files, reading the contents into memory can be impractical. So it is best to use the file object itself as an iterator, which will consume content from the file as needed with no intermediary memory requirement.

```python
efe=open('colores.txt','r')
for x in efe:
    print x,
efe.close()
```

Characters can be read with the read() method.

```python
efe=open('colores.txt','r')
while True:
    ch=efe.read(1)
    if ch!="\n":
        print ch
    if ch == '':
        break
efe.close()
```

One can read more than one character at a time.

```python
efe=open('colores.txt','r')
while True:
    ch=efe.read(2)
    if ch!="\n":
        print ch
    if ch == '':
        break
efe.close()
```

If read is used with no arguments, the entire file is read.

```python
efe=open('colores.txt','r')
print efe.read()
efe.close()
```

***open()***  returns a file object, and is most commonly used with two arguments: open(filename, mode).

The first argument is a string containing the filename. The second argument is another string containing a few characters describing the way in which the file will be used. mode can be 'r' when the file will only be read, 'w' for only writing (an existing file with the same name will be erased), and 'a' opens the file for appending; any data written to the file is automatically added to the end. 'r+' opens the file for both reading and writing. The mode argument is optional; 'r' will be assumed if it’s omitted.

It is good practice to use the ***with*** keyword when dealing with file objects. This has the advantage that the file is properly closed after its suite finishes, even if an exception is raised on the way. It is also much shorter than writing equivalent try-finally blocks:

```python
with open('colores.txt') as f:
    thecolors=f.read()
print thecolors
```

## Valor medio y desviación estándar

```python
randnums=random.randn(100000)
# Desviación estándar
print std(randnums)
# El promedio
print mean(randnums)
# La mediana
print median(randnums)
aver=mean(randnums)
devi=std(randnums)
# contar
bra1=0
bra2=0
bra3=0
for n in randnums:
    if abs(n-aver)<devi:
        bra1+=1
    if abs(n-aver)<2*devi:
        bra2+=1
    if abs(n-aver)<3*devi:
        bra3+=1
# En una distribución normal el 68% de los datos están a una desviación
# estándar de la media,
print float(bra1)/len(randnums)*100
# 95% están a dos  desviaciones estándar,
print float(bra2)/len(randnums)*100
# y el 99.7% están a tres desviaciones estándar.
print float(bra3)/len(randnums)*100
figure(figsize=(10,5))
hist(randnums,10,range=(-3,-2),color="cyan")
hist(randnums,10,range=(-2,-1),color="red")
hist(randnums,20,range=(-1,1),color="blue")
hist(randnums,10,range=(1,2),color="red")
hist(randnums,10,range=(2,3),color="cyan")
xlim(-3,3)
title(u"Distribución normal")
show()
```

En una distribución de Poisson el promedio es igual al cuadrado de la desviación estándar: 
$\bar{x} = \sigma^2$

```python
# Posiciones de estrellas de magnitud menor o igual a 6
estrellas = genfromtxt(os.path.expanduser("~/Dropbox/Data/starpositions.csv"),delimiter=",")
print "Número de estrellas = "+str(len(estrellas))
```

```python
figure(figsize=(18,9))
plot(estrellas[:,0],estrellas[:,1],"o",ms=2)
xlim(0,360)
ylim(-90,90)
grid(b=True)
xlabel(u"Ascención Recta")
ylabel(u"Declinación")
show()
```

```python
Image(filename=os.path.expanduser("~/Dropbox/Data/gra.png"))
```

```python
x=estrellas[:,0]
bucket=zeros(3000)
for star in x:
    i=math.floor(star/(360./3000.))
    bucket[i]+=1
```

```python
figure(figsize=(18,9))
prom=mean(bucket)
print "Estrellas promedio = " + str(prom)
x=arange(0,8)
y=[prom**i/math.factorial(i)*exp(-prom) for i in x]
hist(bucket.flatten(),7,range=(0,7),normed=True,align='left',rwidth=0.5,label="Estrellas")
hist(x,7,weights=y,rwidth=0.5,label="Poisson")
xlim(-0.5,7)
legend()
xlabel(u"Estrellas contadas")
ylabel("Frecuencia relativa")
show()
```

### Luz Coherente

```python
laser10=genfromtxt(os.path.expanduser("~/Dropbox/Data/laser10mu.csv"))
```

```python
figure(figsize=(18,9))
prom=mean(laser10)
sigm=std(laser10)
print "Conteos promedio (p) = " + str(prom)
print "Desviación estándar (s) = " + str(sigm)
print "s^2 = " + str(sigm**2) + " = " + str(prom) + " = " + "promedio"
x=arange(0,5)
y=[prom**i/math.factorial(i)*exp(-prom) for i in x]
hist(laser10,5,range=(0,5),normed=True,align='left',rwidth=0.5,label="Fotones")
hist(x,100,range=(0,100),weights=y,rwidth=0.5,label="Poisson",normed=True)
xlim(-0.5,5)
legend()
xlabel(u"Fotones detectados en 10 $\mu\mathrm{s}$")
ylabel("Frecuencia relativa")
title("Experimento de conteo de fotones \n Luz Coherente (10 $\mu\mathrm{s}$)")
show()
```
