Se usa Runge-Kutta de orden 4 para determinar el movimiento de 3 cuerpos<BR>
Se determina el arreglo `Y`, el cual contiene las comoponentes de *x* y *y* de distancia y velocidad de cada partícula en orden.<BR>
En total, el arreglo `Y` tiene 12 entradas.<BR>

```python
#Se define la funcion f, que recibe el vector Y y arroja su derivada, es decir velocidad y aceleración.
def f(Y):
    r12=((Y[4]-Y[0])**2+(Y[5]-Y[1])**2)**(3/2.)
    r13=((Y[8]-Y[0])**2+(Y[9]-Y[1])**2)**(3/2.)
    r23=((Y[8]-Y[4])**2+(Y[9]-Y[5])**2)**(3/2.)
    list=[]
    list.append(Y[2])
    list.append(Y[3])
    list.append((Y[4]-Y[0])/r12 + (Y[8]-Y[0])/r13)
    list.append((Y[5]-Y[1])/r12 + (Y[9]-Y[1])/r13)
    list.append(Y[6])
    list.append(Y[7])
    list.append((Y[0]-Y[4])/r12 + (Y[8]-Y[4])/r23)
    list.append((Y[1]-Y[5])/r12 + (Y[9]-Y[5])/r23)
    list.append(Y[10])
    list.append(Y[11])
    list.append((Y[0]-Y[8])/r13 + (Y[4]-Y[8])/r23)
    list.append((Y[1]-Y[9])/r13 + (Y[5]-Y[9])/r23)
    arreglo=np.array(list)
    return arreglo
```

```python
#Se define la función que realiza una iteración de Runge-Kutta de 4to orden
def one_Kutta4_step():
    global t
    global Deltat
    global Y
    K1 = f(Y)
    K2 = f(Y + Deltat*(1./2.*K1))
    K3 = f(Y + Deltat*(1./2.*K2))
    K4 = f(Y + Deltat*K3)
    Y+= Deltat * (1/6.*K1 + 1/3. * K2 + 1/3.*K3 + 1/6.*K4)
    t+= Deltat
```

```python
chosenstep=one_Kutta4_step
numIter=100
a=0.24250109
b=-0.0607718825
c=0.93240737
d=0.86473146
Y=np.array([a,b,c,d,-a,-b,c,d,0,0,-2*c,-2*d])
tmax=0.5
t=0.
Deltat=tmax/numIter
history=np.array([[t,Y[0],Y[1],Y[2],Y[3],Y[4],Y[5],Y[6],Y[7],Y[8],Y[9],Y[10],Y[11]]])
for __ in range(numIter):
    chosenstep()
    history=np.append(history,[[t,Y[0],Y[1],Y[2],Y[3],Y[4],Y[5],Y[6],Y[7],Y[8],Y[9],Y[10],Y[11]]],axis=0)
plt.figure(figsize=(10,5))
plt.plot(history[:,1],history[:,2])
plt.plot(history[-1,1],history[-1,2],"o")
plt.title("Trayectoria de m1")
plt.show()
```
