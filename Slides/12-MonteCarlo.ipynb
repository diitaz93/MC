## Calculando pi

Volumen de la n-esfera


$V_n(R)=\frac{\pi^{n/2}}{\Gamma(1+n/2)} R^n$

```python
# En 2D
darts=100000
cuenta=0
adentro=np.zeros((darts,2))
afuera=np.zeros((darts,2))
for i in xrange(darts):
    x=2*np.random.random()-1
    y=2*np.random.random()-1
    if x**2+y**2<=1.:
        cuenta+=1
        adentro[i]=np.array([x,y])
    else:
        afuera[i]=np.array([x,y])
print "pi = " + str(4.*cuenta/darts)
```

```python
def carlopi(N):
    lanzamientos=(np.random.random((N,2))-0.5)*2
    inside=((lanzamientos[:,0]**2+lanzamientos[:,1]**2)<=1).sum()
    return 4*float(inside)/N
```

```python
# ¿Cómo escala el error?
pies=np.zeros((8,2))
for i in range(0,8):
    print i,
    elene=5**(i+1)
    pisum=0.
    for j in range(500):
        pisum+=carlopi(elene)
    pies[i-1]=np.array([elene,abs(pisum/500.-np.pi)])
```

```python
plt.figure(figsize=(10,6))
plt.plot(pies[:,0],pies[:,1],"o")
plt.xscale('log')
plt.yscale('log')
plt.xlabel("N")
plt.ylabel("Err")
plt.text(1e3,1e-2,r"$\propto 1/\sqrt{N}$",\
         horizontalalignment='center',fontsize=20)
plt.show()
np.polyfit(np.log(pies[:,0]),np.log(pies[:,1]),1)
```

## Bingo

N personas juegan Bingo, ¿cuál es la probabilidad de que el ganador se obtenga en el turno n?

<img src="http://bingoonlinetop.com/wp-content/uploads/2010/05/15a3.jpg">

```python
# Produce una lista de c números enteros aleatorios en el intervalo [a,b-1] sin que se repitan
def intsample(a,b,c):
    done=0
    nums=[]
    if b-a < c:
        return 0
    while len(nums)<c:
        num=np.random.randint(a,b)
        if not(num in nums):
            nums.append(num)
    return nums
            
def carton():
    bcol=intsample(1,16,5)
    icol=intsample(16,31,5)
    ncol=intsample(31,46,4)
    gcol=intsample(46,61,5)
    ocol=intsample(61,76,5)
    return bcol+icol+ncol+gcol+ocol
```

```python
# Un solo juego
jugadores=10
gano=0
bolsa=range(1,76) # Lista de números por sacar
juego=[0 for i in range(jugadores)] # Lista que guarda el número de aciertos de cada jugador
# Generar una lista con todos los cartones
cartones=[]
numcart=0
while numcart < 10:
    elcarton=carton()
    if elcarton not in cartones:
        cartones.append(carton())
        numcart+=1
turno=0
while gano==0:
    salio=np.random.choice(bolsa)
    if len(bolsa)!=0:
        bolsa.remove(salio)
    for k in range(jugadores): # Actualizar el número de aciertos de cada jugador
        if salio in cartones[k]:
            juego[k]+=1
    if 24 in juego:
        gano=1
    turno+=1
```
