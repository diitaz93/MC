```python
%pylab inline
import numpy
import scipy.optimize as optimization
from scipy.optimize import curve_fit
```

```python
x=numpy.array([2.3,2.8,3.2,3.7,4.3])
y=numpy.array([34745,19689,12594,7982,5822])
```

```python
def func(x,a, m):
    return (a*m)/(x**3)
fitpars, covmat=curve_fit(func,x,y)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y,"ro",ms=10,label="datos")
ax.plot(x,func(x,*fitpars),'b--',lw=2.0,label=u"Interpolación")
plt.xlim(1,5)
plt.ylim(0,40000)
plt.legend(loc='lower left')
plt.title(u'Interpolación Mínimos Cuadrados')
plt.show(fig)
plt.savefig("a.png")
```
