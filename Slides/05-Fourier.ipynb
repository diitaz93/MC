## Series de Fourier

Si f(t) es una función periódica ($f(t+T) = f(t)$) con las siguientes propiedades (condiciones de Dirichlet)

   + $f(t)$ es absolutamente integrable en un periodo,
   + tiene un número finito de máximos y mínimos en un intervalo de un periodo,
   + ,tiene un número finito de discontinuidades en un intervalo finito, sin ser infinitas.
   + y además es una función acotada,

entonces

   $f(t)=\frac{a_0}{2} + \sum_{n=1}^\infty \left( a_n \cos n\omega t + b_n \sin b\omega t \right)$

con 

   $\omega = \frac{2\pi}{T}$
   
   $a_n = \frac{2}{T}\int_{0}^{T} f(t) \cos n \omega t \,\textrm{d}t$

   $b_n = \frac{2}{T}\int_{0}^{T} f(t) \sin n \omega t \,\textrm{d}t$.

**Ejemplo: onda cuadrada**

```python
fig=plt.figure(figsize=(10,5))
x=np.linspace(0,2*np.pi,100)
y=0.*x
plt.plot([0,0,np.pi,np.pi,2*np.pi,2*np.pi],[0,1,1,-1,-1,0],"k",lw=3)
for i in range(5):
    y+=4./np.pi*1./(2*i+1)*np.sin((2*i+1)*x)
    plt.plot(x,y,label=str(i))
plt.xlim(0,2*np.pi)
plt.legend()
plt.title("Square Wave Fourier Series")
plt.show()
```

## Transformada de Fourier Discreta

```python
from scipy.fftpack import ifft, fft, fftfreq
```

```python
help(fftfreq)
```

```python
N = 2**10 # Número de puntos de muestreo
f = 100.0 # Frecuencia
dt = 1./f / N # tiempo entre puntos de muestreo
t = np.linspace(0,1/f,N)
# Tres frecuencias: 100 Hz, 200 Hz y 400 Hz
y = np.sin(2 * np.pi * f * t) + 0.5 * np.sin(2*2 * np.pi * f * t) + 0.25 * np.sin(4*2 * np.pi * f * t)
plt.figure(figsize=(10,5))
plt.plot(t,y,"b",lw=2,label=u'señal')
plt.xlabel("t/s")
plt.legend()
plt.show()
```

```python
# Ahora hagamos la FFT
ffty=fft(y)
freq = fftfreq(N, dt) # frecuencias
absffty=np.abs(ffty)
plt.figure(figsize=(10,5))
for i in range(len(absffty)):
    plt.plot([freq[i],freq[i]],[0,absffty[i]],"-")
plt.plot(freq,absffty,"o")
plt.xlabel("f/Hz")
plt.xlim(-500,500)
plt.title("DFT")
plt.show()
# La DFT muestra claramente las frecuencias presentes en la señal
```

## Filtrado de Señales

```python
N = 2**10 # Número de puntos de muestreo
f = 100.0 # Frecuencia
dt = 1./f / N # tiempo entre puntos de muestreo
t = np.linspace(0,1/f,N)
y = np.sin(2 * np.pi * f * t) + 0.2 * np.cos(2*2 * np.pi * f * t) + 0.2 * np.cos(5*2 * np.pi * f * t) # Esta es la señal
ruido = 0.2*(np.random.randn(N)) # y este el ruido.
y = y + ruido
plt.figure(figsize=(10,5))
plt.plot(t,y,"r",alpha=0.7,label=u'señal ruidosa')
plt.plot(t,y-ruido,"b",lw=2,label=u'señal')
plt.xlabel("t/s")
plt.legend()
plt.show()
```

```python
# Ahora calculamos las DFT
fft_x_clean = fft(y-ruido)
fft_x_noisy = fft(y)
freq = fftfreq(N, dt) # frecuencias
plt.figure(figsize=(10,5))
plt.plot(freq,np.abs(fft_x_clean),"o",label=u'DFT señal')
plt.plot(freq,-np.abs(fft_x_noisy),"o",label=u'DFT ruido + señal')
plt.legend()
plt.xlim(-1000,1000)
plt.show()
```

```python
# En el filtro de eliminan las frecuencias altas
plt.figure(figsize=(15,20))
freq_cuts=np.logspace(2,4,10)
for k in range(1,11):
    freq_max = freq_cuts[-k]
    fft_x_noisy[np.abs(freq) > freq_max] = 0
    # Ahora se toma la transformada inversa
    y_cleaned=ifft(fft_x_noisy)
    plt.subplot(5,2,k)
    plt.plot(t,np.real(y_cleaned),label=u'Señal filtrada')
    plt.plot(t,np.real(y-ruido),label=u'Señal original sin ruido')
    plt.legend()
    plt.title("f_cut = "+str(freq_max))
    plt.xlabel("t/s")
plt.tight_layout()
plt.show()
```

## Compresión de Imágenes

```python
from scipy import misc
from scipy.fftpack import ifft, fft, fftfreq, fft2, ifft2, fftn, ifftn
```

```python
# Invitar a Lena
lena=misc.lena()
# Lena es un array
print type(lena)
plt.figure(figsize=(10,10))
plt.imshow(lena,cmap=plt.cm.gray)
plt.show()
```

```python
# Ahora encontremos la transformada de Fourier de Lena
plt.figure(figsize=(15,15))
lena_dft=fft2(lena)
mag_lena_dft=np.log(np.abs(lena_dft))
fx, fy = np.meshgrid(np.linspace(0,1,len(lena)),np.linspace(0,1,len(lena)))
#lena_dft[np.sqrt(fx**2+fy**2) > 0.2]=0.
plt.subplot(2,2,1)
plt.imshow(mag_lena_dft)
lim=0.1
pick=(np.sqrt((fx-1)**2+fy**2) > lim)*(np.sqrt((fx)**2+fy**2) > lim)*(np.sqrt(fx**2+(fy-1)**2) > lim)*(np.sqrt((fx-1)**2+(fy-1)**2) > lim)
lena_dft[pick]=0.
mag_lena_dft=np.log(np.abs(lena_dft))
plt.subplot(2,2,2)
plt.imshow(mag_lena_dft)
angle_lena_dft=np.angle(lena_dft)
lena_recovered=np.real(ifft2(lena_dft))
plt.subplot(2,2,3)
plt.imshow(lena,interpolation=None,cmap=plt.cm.gray)
plt.subplot(2,2,4)
plt.imshow(lena_recovered,interpolation=None,cmap=plt.cm.gray)
plt.show()
```

```python
# Una idea diferente para el filtro
plt.figure(figsize=(10,10))
lena_dft=fft2(lena)
mag_lena_dft=np.log(np.abs(lena_dft))
fx, fy = np.meshgrid(np.linspace(0,1,len(lena)),np.linspace(0,1,len(lena)))
#lena_dft[np.sqrt(fx**2+fy**2) > 0.2]=0.
plt.subplot(2,2,1)
plt.imshow(mag_lena_dft)
lim=10
pick=mag_lena_dft < lim
lena_dft[pick]=0.
mag_lena_dft=np.log(np.abs(lena_dft))
plt.subplot(2,2,2)
plt.imshow(mag_lena_dft)
angle_lena_dft=np.angle(lena_dft)
lena_recovered=np.real(ifft2(lena_dft))
plt.subplot(2,2,3)
plt.imshow(lena,interpolation=None,cmap=plt.cm.gray)
plt.subplot(2,2,4)
plt.imshow(lena_recovered,interpolation=None,cmap=plt.cm.gray)
plt.show()
```
