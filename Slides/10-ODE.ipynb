## Método de Euler y Regla del Punto Medio

$$Y(t_{n+1}) \approx Y(t_n) + f(Y(t_n),t_n) \Delta t$$
$$\textrm{Forward}$$

```python
def f(z):
    return 1.j*z
def one_Euler_step():
    global t
    global Deltat
    global z
    z+= f(z)*Deltat
    t+=Deltat
```

```python
numIters=np.logspace(1,3,20,2)
lastones=[]
plt.figure(figsize=(7,7))
plt.subplot(211)
chosenstep=one_Euler_step
for numIterf in numIters:
    z=1.
    tmax=2*np.pi
    t=0.
    numIter=np.int(numIterf)
    Deltat=tmax/numIter
    history=[[t,z]]
    for __ in range(numIter):
        chosenstep()
        history.append([t,z])
    ahistory=np.array(history)
    lastones.append(history[-1])
    plt.plot(np.real(ahistory[:,1]),np.imag(ahistory[:,1]),label=r"$\Delta t = $"+str(np.round(Deltat,3)))
plt.legend()
plt.title("Euler Method")
plt.axis('equal')
alastones=np.array(lastones)
plt.subplot(212)
plt.plot(tmax/numIters,np.abs(1-alastones[:,1]),"o")
plt.xscale('log')
plt.yscale('log')
plt.show()
np.polyfit(np.log(tmax/numIters),np.log(np.abs(1-alastones[:,1])),1)
```

$$Y(t_{n+1}) \approx Y(t_n) + \frac{1}{2}(K_1+K_2)\Delta t$$
$$ K_1 = f(t_n)$$
$$ K_2 = f( Y + \frac{df}{dY} * f(t_n) \Delta t$$
$$\textrm{Midpoint Rule}$$

```python
def fp(z):
    return 1.j
def one_midpoint_step():
    global t
    global Deltat
    global z
    K1= f(z)
    K2= f(z + f(z)*Deltat) 
    z+= Deltat/2. * (K1+K2)
```

```python
numIters=np.logspace(1,3,20,2)
lastones=[]
plt.figure(figsize=(7,7))
plt.subplot(211)
chosenstep=one_midpoint_step
for numIterf in numIters:
    z=1.
    tmax=2*np.pi
    t=0.
    numIter=np.int(numIterf)
    Deltat=tmax/numIter
    history=[[t,z]]
    for __ in range(numIter):
        chosenstep()
        history.append([t,z])
    ahistory=np.array(history)
    lastones.append(history[-1])
    plt.plot(np.real(ahistory[:,1]),np.imag(ahistory[:,1]),label=r"$\Delta t = $"+str(np.round(Deltat,3)))
plt.legend()
plt.title("Midpoint rule")
plt.axis('equal')
alastones2=np.array(lastones)
plt.subplot(212)
plt.plot(tmax/numIters,np.abs(1-alastones2[:,1]),"o")
plt.xscale('log')
plt.yscale('log')
plt.show()
np.polyfit(np.log(tmax/numIters),np.log(np.abs(1-alastones2[:,1])),1)
```

## Runge-Kutta 3er orden

```python
def one_Kutta3_step():
    global t
    global Deltat
    global z
    K1 = f(z)
    K2 = f(z+Deltat*(1/2.*K1))
    K3 = f(z + Deltat*(-K1+2.*K2))
    z+= Deltat / 6. * (K1+4*K2+K3)
```

```python
numIters=np.logspace(1,3,20,2)
lastones=[]
plt.figure(figsize=(7,7))
plt.subplot(211)
chosenstep=one_Kutta3_step
for numIterf in numIters:
    z=1.
    tmax=2*np.pi
    t=0.
    numIter=np.int(numIterf)
    Deltat=tmax/numIter
    history=[[t,z]]
    for __ in range(numIter):
        chosenstep()
        history.append([t,z])
    ahistory=np.array(history)
    lastones.append(history[-1])
    plt.plot(np.real(ahistory[:,1]),np.imag(ahistory[:,1]),label=r"$\Delta t = $"+str(np.round(Deltat,3)))
plt.legend()
plt.title("Kutta Orden 3")
plt.axis('equal')
alastones3=np.array(lastones)
plt.subplot(212)
plt.plot(tmax/numIters,np.abs(1-alastones3[:,1]),"o")
plt.xscale('log')
plt.yscale('log')
plt.show()
np.polyfit(np.log(tmax/numIters),np.log(np.abs(1-alastones3[:,1])),1)
```

## Runge-Kutta Orden 4

```python
def one_Kutta4_step():
    global t
    global Deltat
    global z
    K1 = f(z)
    K2 = f(z + Deltat*(1/2.*K1))
    K3 = f(z + Deltat*(1./2.*K2))
    K4 = f(z + Deltat*K3)
    z+= Deltat * (1/6.*K1 + 1/3. * K2 + 1/3.*K3 + 1/6.*K4)
```

```python
numIters=np.logspace(1,3,20,2)
lastones=[]
plt.figure(figsize=(7,7))
plt.subplot(211)
chosenstep=one_Kutta4_step
for numIterf in numIters:
    z=1.
    tmax=2*np.pi
    t=0.
    numIter=np.int(numIterf)
    Deltat=tmax/numIter
    history=[[t,z]]
    for __ in range(numIter):
        chosenstep()
        history.append([t,z])
    ahistory=np.array(history)
    lastones.append(history[-1])
    plt.plot(np.real(ahistory[:,1]),np.imag(ahistory[:,1]),label=r"$\Delta t = $"+str(np.round(Deltat,3)))
plt.legend()
plt.title("Kutta Orden 4")
plt.axis('equal')
alastones4=np.array(lastones)
plt.subplot(212)
plt.plot(tmax/numIters,np.abs(1-alastones4[:,1]),"o")
plt.xscale('log')
plt.yscale('log')
plt.show()
np.polyfit(np.log(tmax/numIters),np.log(np.abs(1-alastones4[:,1])),1)
```

```python
plt.figure(figsize=(10,5))
plt.plot(tmax/numIters,np.abs(1-alastones4[:,1]),"o--",label="Orden 2")
plt.plot(tmax/numIters,np.abs(1-alastones3[:,1]),"o--",label="Orden 3")
plt.plot(tmax/numIters,np.abs(1-alastones2[:,1]),"o--",label="Orden 4")
plt.xscale('log')
plt.yscale('log')
plt.xlabel('step size') 
plt.ylabel('error')
plt.legend()
plt.show()
```

## El problema de Kepler

```python
def f(Y):
    r3=(np.sqrt(Y[0]**2+Y[1]**2))**3
    return np.array([Y[2],Y[3],-Y[0]/r3,-Y[1]/r3])
```

```python
def one_Kutta4_step():
    global t
    global Deltat
    global Y
    K1 = f(Y)
    K2 = f(Y + Deltat*(1./2.*K1))
    K3 = f(Y + Deltat*(1./2.*K2))
    K4 = f(Y + Deltat*K3)
    Y+= Deltat * (1/6.*K1 + 1/3. * K2 + 1/3.*K3 + 1/6.*K4)
    t+= Deltat
```

```python
chosenstep=one_Kutta4_step
numIter=100
Y=np.array([1.,0.,0.,1.])
tmax=2*np.pi
t=0.
Deltat=tmax/numIter
history=np.array([[t,Y[0],Y[1],Y[2],Y[3]]])
for __ in range(numIter):
    chosenstep()
    history=np.append(history,[[t,Y[0],Y[1],Y[2],Y[3]]],axis=0)
plt.figure()
plt.plot(history[:,1],history[:,2],"o",ms=1)
plt.title("Kutta Orden 4")
plt.axis('equal')
plt.show()
```
