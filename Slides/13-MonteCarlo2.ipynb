## Importance Sampling

```python
HTML('<iframe src=https://en.wikipedia.org/wiki/Monte_Carlo_integration#Importance_sampling_algorithm width=800 height=300></iframe>')
```

$\int_{-4}^{4}\int_{-4}^{4}\int_{-4}^{4}\int_{-4}^{4} (x^2+y^2+z^2+w^2) \frac{1}{4\pi^2}e^{-(x^2+y^2+z^2+w^2)/2} \textrm{d}x\textrm{d}y\textrm{d}z\textrm{d}w$

```python
def integrando(x,y,z,w):
    return (x**2+y**2+z**2+w**2)*np.exp(-(x**2+y**2+z**2+w**2)/2)/4./np.pi**2
```

```python
# Con números tomados de una distribución uniforme
darts=10000
intentos=1000
resultados_u=[]
for __ in xrange(intentos):
    resultados_u.append(8.**4/darts*sum(integrando(np.random.uniform(-4.,4.,darts),np.random.uniform(-4,4,darts),np.random.uniform(-4,4,darts),np.random.uniform(-4,4,darts))))
```

```python
def gauss4(x,y,z,w):
    return np.exp(-(x**2+y**2+z**2+w**2)/2)/4./np.pi**2
def integrand_factor(x,y,z,w):
    return (x**2+y**2+z**2+w**2)
darts=10000
intentos=1000
resultados_p=[]
# Con números tomados con una distribución de la misma gaussiana
for __ in xrange(intentos):
    randx=np.random.randn(darts)
    randy=np.random.randn(darts)
    randz=np.random.randn(darts)
    randw=np.random.randn(darts)
    resultados_p.append(1./darts*sum(integrand_factor(randx,randy,randz,randw)))
```

```python
with plt.xkcd():
    plt.figure(figsize=(10,8))
    plt.subplot(2,1,1)
    plt.hist(resultados_u,normed=True,label='Muestreo uniforme')
    plt.legend()
    plt.xlim(3.4,4.6)
    plt.subplot(2,1,2)
    plt.hist(resultados_p,normed=True,color="red",label='Importance Sampling')
    plt.legend()
    plt.xlim(3.4,4.6)
    plt.text(4.2,6,"Reduced variance!")
    plt.show()
```

$f(x,y)=\frac{1}{2\pi} e^{-(x**2+y**2)/2}$

```python
def gauss2(x,y):
    return np.exp(-(x**2+y**2)/2)/2./np.pi
steps=5000
stepsize=1.
x=np.random.uniform(-10,10)
y=np.random.uniform(-10,10)
history1=np.array([[x,y]])
for __ in xrange(steps):
    phi=2.*np.pi*np.random.random()
    dx=stepsize*np.cos(phi)
    dy=stepsize*np.sin(phi)
    gi=gauss2(x,y)
    gf=gauss2(x+dx,y+dy)
    if gf > gi :
        x+=dx
        y+=dy
    elif np.random.random()<min(1,gf/gi):
        x+=dx
        y+=dy
    history1=np.append(history1,[[x,y]],axis=0)

x=np.random.uniform(-10,10)
y=np.random.uniform(-10,10)
history2=np.array([[x,y]])
for __ in xrange(steps):
    phi=2.*np.pi*np.random.random()
    dx=stepsize*np.cos(phi)
    dy=stepsize*np.sin(phi)
    gi=gauss2(x,y)
    gf=gauss2(x+dx,y+dy)
    if gf > gi :
        x+=dx
        y+=dy
    elif np.random.random()<min(1,gf/gi):
        x+=dx
        y+=dy
    history2=np.append(history2,[[x,y]],axis=0)

x=np.random.uniform(-10,10)
y=np.random.uniform(-10,10)
history3=np.array([[x,y]])
for __ in xrange(steps):
    phi=2.*np.pi*np.random.random()
    dx=stepsize*np.cos(phi)
    dy=stepsize*np.sin(phi)
    gi=gauss2(x,y)
    gf=gauss2(x+dx,y+dy)
    if gf > gi :
        x+=dx
        y+=dy
    elif np.random.random()<min(1,gf/gi):
        x+=dx
        y+=dy
    history3=np.append(history3,[[x,y]],axis=0)
```

## Modelo de Ising

$$H=-\sum_{<i,j>} x_ix_j$$

$$M=\sum{x_i}$$

```python
N=64
lattice=np.ones((N,N))
T=10.
beta=1./T
def total_energy():
    global lattice
    return -sum(sum(lattice*(np.roll(lattice,1,axis=1)+\
                         np.roll(lattice,-1,axis=1)+\
                         np.roll(lattice,1,axis=0)+\
                         np.roll(lattice,-1,axis=0))))
def block_energy(i,j):
    global lattice
    global N
    return -(lattice[i,j]*(lattice[(i-1)%N,j]+\
                                  lattice[(i+1)%N,j]+\
                                  lattice[i,(j+1)%N]+\
                                  lattice[i,(j-1)%N]))
def flipone():
    global lattice
    global beta
    global N
    # Pick a random site
    nx, ny= np.random.randint(0,N),np.random.randint(0,N)
    # Calculate current energy of site and neighbors
    en_i=block_energy(nx,ny)
    # If it were flipped
    en_f=-en_i
    if en_f<en_i: # This could be simplified
        lattice[nx,ny]*=-1
    elif np.random.random()<min(1,np.exp(-beta*(en_f-en_i))):
        lattice[nx,ny]*=-1

def magnetization():
    global lattice
    global N
    return sum(sum(lattice))/N**2
```
