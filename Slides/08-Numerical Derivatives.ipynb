**Forward, backward and central differences**

$$f'(x) = \frac{f(x+h)-f(x)}{h} + O(h)$$

$$\textrm{(Foward difference)}$$

$$f'(x) = \frac{f(x)-f(x-h)}{h} + O(h) $$

$$\textrm{(Backward difference)}$$

$$f'(x) = \frac{f(x+h)-f(x-h)}{2h} + O(h^2)$$

$$\textrm{(Central difference)}$$

```python
# Ejemplo de derivadas por diferencias futuras, pasadas y divididas

k=1.
def examplefun(x):
    return np.sin(k*x)
def examplefunderivative(x):
    return k*np.cos(k*x)
numPoints=15
a=0. # Límite inferior
b=2*np.pi # Límite superior
finex=np.linspace(a,b,numPoints*10) # Esto es para hacer las gráficas más detalladas
x=np.linspace(a,b,numPoints) # Sobre este array se evalúan las derivadas
y=examplefun(x) # Las abscisas

h = (b-a)/(numPoints-1) # Este es el paso del arreglo x
yforward = np.roll(y,-1) # toma el arreglo y y lo mueve una posición hacia la izquierda
ybackward = np.roll(y,1) # y esto una posición a la derecha

# Calcular las derivadas usando forward differences
dy1 = (yforward-y)/h
dy1 = dy1[:-1] # la última no tiene significado

# ahora usando diferencias pasadas,
dy2=(y-ybackward)/h
dy2=dy2[1:] # la primera no tiene significado

# y finalmente usando diferencias centrales.
dy3 = (yforward-ybackward)/(2.*h)
# La primera y la última no tienen significado
dy3=dy3[1:-1]

#Gráficas
with plt.xkcd():
    plt.figure(figsize=(20,5))
    # plt.plot(finex,examplefun(finex),"--",lw=3,label="Function")
    plt.plot(x[:-1],dy1,"o-",label="Foward differences")
    plt.plot(x[1:],dy2,"o-",label="Backward differences")
    plt.plot(x[1:-1],dy3,"o-",label="Central differences")
    # True derivative
    plt.plot(finex,examplefunderivative(finex),"--",lw=3,label="\"True\" derivative")
    plt.legend()
    plt.xlim(a,b)
    plt.xlabel("x")
    plt.ylabel("f'(x)")
    plt.show()
```

```python
# Estudio numérico del error calculando la derivada en pi/4
xchoice=np.pi/4.
hvals=np.logspace(-10,-1,20)

y = np.sin(xchoice)
truedev=1./np.sqrt(2)
derivadas=[]
for h in hvals:
    yforward = np.sin(xchoice+h) # toma el arreglo y y lo mueve una posición hacia la izquierda
    ybackward = np.sin(xchoice-h) # y esto una posición a la derecha
    # Calcular las derivadas usando forward differences
    dy1 = (yforward-y)/h
    # ahora usando diferencias pasadas,
    dy2=(y-ybackward)/h
    # y finalmente usando diferencias centrales.
    dy3 = (yforward-ybackward)/(2.*h)
    derivadas.append([h,np.abs(dy1-truedev),np.abs(dy2-truedev),np.abs(dy3-truedev)])
aderivadas=np.array(derivadas)

# Ahora hacer gráficas
plt.figure(figsize=(10,5))
plt.plot(aderivadas[:,0],aderivadas[:,1],"o-",label='Diferencia futura')
plt.plot(aderivadas[:,0],aderivadas[:,2],"o-",label='Diferencia pasada')
plt.plot(aderivadas[:,0],aderivadas[:,3],"o-",label='Diferencia central')
#plt.xlim(-0.1,0.1)
plt.xlabel("Step size (h)")
plt.ylabel("err")
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.show()
```

```python
from scipy import interpolate as interscipy
```

```python
from sympy import *
x,y,z = symbols('x y z')
init_printing(use_unicode=True)
```

$$D(h) = \frac{f(x+h)-f(x-h)}{2h} $$

```python
D0,D1 = symbols('D0 D1')
h0, h1= symbols('h0 h1')
h1 = h0/2
lag=D0*(x**2-h1**2)/(h0**2-h1**2) + D1*(x**2-h0**2)/(h1**2-h0**2)
lag.subs(x,0)
```

```python
D0,D1,D2 = symbols('D0 D1 D2')
h0, h1, h2= symbols('h0 h1 h2')
h1 = h0/2
h2 = h0/4
lag=D0*(x**2-h1**2)*(x**2-h2**2)/(h0**2-h1**2)/(h0**2-h2**2) + D1*(x**2-h0**2)*(x**2-h2**2)/(h1**2-h0**2)/(h1**2-h2**2) + D2*(x**2-h0**2)*(x**2-h1**2)/(h2**2-h0**2)/(h2**2-h1**2)
lag.subs(x,0)
```

```python
def D(x,h):
    return (examplefun(x+h)-examplefun(x-h))/(2.*h)
Dtable=[]
h=0.01/2.
xchoice=np.pi/4.
for i in range(5):
    Dtable.append([h,D(xchoice,h)])
    h=h/2.
aDtable=np.array(Dtable)
finex=np.linspace(0,0.01/2,10)
plt.figure(figsize=(10,5))

plt.xlabel(r"$h$")
plt.ylabel(r"$D$")
# Calcular ahora una interpolación polinomial de orden 1 
# en h^2 usand D_0 y D_1
# y la usamos para extrapolar D(0).
rows=aDtable[0:2]
lag1=interscipy.lagrange(rows[:,0]**2,rows[:,1])
plt.plot(finex,lag1(finex**2),"c-",lw=3,label="Lagrange orden 1 en h^2")
# Calcular ahora una interpolación polinomial de orden 2 en h^2 usand D_0, D_1 y D2,
# y usarla para extrapolar.
rows=aDtable[0:3]
lag2=interscipy.lagrange(rows[:,0]**2,rows[:,1])
plt.plot(finex,lag2(finex**2),"--",lw=3,label="Lagrange orden 2 en h^2")
plt.plot(aDtable[:,0],aDtable[:,1],"or")
plt.legend()
plt.show()
```

```python
# Estudio numérico del error calculando la derivada en pi/4
xchoice=np.pi/4.
hvals=np.logspace(-7,0,20)

y = np.sin(xchoice)
truedev=1./np.sqrt(2)
derivadasC=[]

for h in hvals:
    yforward = np.sin(xchoice+h) # toma el arreglo y y lo mueve una posición hacia la izquierda
    ybackward = np.sin(xchoice-h) # y esto una posición a la derecha
    # Calcular usando diferencias centrales.
    dyc = (yforward-ybackward)/(2.*h)
    derivadasC.append([h,np.abs(dyc-truedev)])

aderivadasC=np.array(derivadasC)

derivadasE1=[]
derivadasE2=[]

for h in hvals:
    D0=D(xchoice,h)
    D1=D(xchoice,h/2.)
    D2=D(xchoice,h/4.)
    derivadasE1.append([h,np.abs(truedev-(-1/3. * D0 + 4./3 * D1))])
    derivadasE2.append([h,np.abs(truedev-(1/45. * D0 - 4./9 * D1 + 64./45 * D2))])

aderivadasE1=np.array(derivadasE1)
aderivadasE2=np.array(derivadasE2)
    
# Ahora hacer gráficas
plt.figure(figsize=(10,5))
plt.plot(aderivadasC[:,0],aderivadasC[:,1],"o-",label='Diferencia central')
plt.plot(aderivadasE1[:,0],aderivadasE1[:,1],"o-",label=u'Extrapolación de Orden 1')
plt.plot(aderivadasE2[:,0],aderivadasE2[:,1],"o-",label=u'Extrapolación de Orden 2')
#plt.xlim(-0.1,0.1)
plt.xlabel("Step size (h)")
plt.ylabel("err")
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.show()
```
